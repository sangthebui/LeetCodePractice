## Sprague-Grundy Theorem
- Same moves available to both players
- No randomness
- No draws
- Last move wins

- Grundy Number
    - For a position G, its Grundy number is: g(G) = mex{g(G1), g(G2)...}
    - where G1, G2 are all positions reachable in 1 move.
    - mex: minimum excluded non-integer integer

    - e.g. if reachable Grundy numbers are {0, 1, 3}, then mex = 2

- Base Case
    - If a position has no legal moves: g(G) = 0
    - This is a losing position

- Key theorem
    - g(G1 + G2 + ..Gn) = g(G1) ^ g(G2) ^ g(Gn)
    - ^ bitwise XOR

- Generic code template for DP
```python
def grundy(state):
    if state in memo:
        return memo[state]

    reachable = set()
    for next_state in moves(state):
        reachable.add(grundy(next_state))

    g = mex(reachable)
    memo[state] = g
    return g
```
