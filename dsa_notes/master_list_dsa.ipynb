## 3. Two Pointers
- generic algorithm of 2 Pointers
- usually also need a target -> find something

```python
# binary search
nums = [1,2,3,4, 5, 6, 7]
target = 3
def binary_search(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = left + ((right - left) // 2)
        if nums[mid] == target:
            return mid
        elif nums[mid] > target:
            right = mid - 1
        else:
            left = mid + 1
    return -1
```
