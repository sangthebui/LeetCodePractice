### Binary Search - sorted array
- problems with a sorted array and we need to search through it.

````python
# binary search if array is sorted
nums = [1,2,3,4, 5, 6, 7]
target = 3
def binary_search(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = left + ((right - left) // 2)
        if nums[mid] == target:
            return mid
        elif nums[mid] > target:
            right = mid - 1
        else:
            left = mid + 1
    return -1
````

### Binary Search with bisect
- Use bisect module to simplify binary search

````python
import bisect

def binary_search_with_bisect(nums, target):
    i = bisect.bisect_left(nums, target)
    # Check two conditions to confirm the target is actually in the list:
    # 1. 'i' must be a valid index (not equal to the length of the list)
    # 2. The element at index 'i' must exactly match the target value
    return i if i != len(nums) and nums[i] == target else -1

# could aura farm with this 1 line binary search with if comprehension and walrus operator
def binary_search_with_bisect_oneliner(nums, target):

    return i if (i := bisect.bisect_left(nums, target)) != len(nums) and nums[i] == target else -1
````

### Palindrome and Reverse a string

````python
# palindrome, reverse a string,
# problems where the two pointers are at opposite end and meet in the middle
def reverse_string(a_string):
    left = 0
    right = len(str) - 1
    while left < right:
        # swap them for reverse, check for same if palindrome
        a_string[left], a_string[right] = a_string[right], a_string[left]
        left += 1
        right -= 1
````

### Greedy method - Find max
- Find the max by taking all the arrays first, then shrink left or right unless they converge.
- example: Container With Most Water

````python
# given array of heights, find the max area. 
# goal is to use the left and right as width and mutiply it by the smaller of the two heights (left and right)
def max_area(height):
    left = 0
    right = len(height) - 1
    max_area = 0
    while left < right:
        area = (right - left) * min(height[right], height[left])
        max_area = max(max_area, area)

        if (height[left] < height[right]):
            left += 1
        else:
            right -= 1
    return max_area
````

### Floyd's Tortoise and Hare (Fast and Slow Pointers)
- used to detect cycles for problems with repeating substructure
- could be use to detect cycles in a Linked Lists, or a problem with modulo arithemic's nature of loops.
- problems where if we keep doing something, it will loop around

````python
# detect a happy  number
# Starting with any positive integer, replace the number by the sum of the squares of its digits.
# Repeat the process until the number equals 1 (where it will stay), or it loops endlessly in a cycle which does not include 1.
# Those numbers for which this process ends in 1 are happy.
def next_number(n: int) -> int:
    return sum(int(digit)**2 for digit in str(n))


def isHappy(n: int) -> bool:
    slow = n
    fast = n
    while True:
        slow = next_number(slow)
        fast = next_number(next_number(fast))
        if slow == fast:
            break
    return slow == 1
````

### Sliding window is the sum

````python
# generic code for a sliding window
# Maximm average subarray I
# Find a contiguous subarray whose length is equal to k that has 
# the maximum average value and return this value. 
# Any answer with a calculation error less than 10-5 will be accepted.
def find_max_average(nums, k):
    window = sum(nums[:k])
    maximum = window 
    for right in range(k, len(nums)):
        left = right - k
        new_window = window - nums[left] + nums[right]
        maximum = max(maximum, new_window)
    return maximum / k 

# this sliding window is the average sum of all the numbers.
# the window size is k
# the constraint of the problem is that the number is small enough where it won't overflow
# could just add them all up and them get the average at the end.
````

### Sliding window is a set


````python
# the problem Contain Duplicate II
# Given an integer nums and an integer k, return true if there are 2 distinct indices i and j
# in the array such nums[i] == nums[j] and abs(i - j) <= k
def contain_duplicate_2(nums, k):
    window = set()
    left = 0
    for right in range(len(nums)):
        num = nums[right]
        if right - left > k:
            window.remove(nums[left])
            left += 1
        
        if num in window:
            return True

        window.add(num)

    return False

# The window is a set of non duplicate numbers 
# and the size of the window is less than or equal to k
# once we violate the size of window, then we remove the left number
````

### Sliding window is a formula

````python
# Longest Repeating Character Replacement
# You are given a string s and an integer k. 
# You can choose any character of the string and 
# change it to any other uppercase English character. 
# You can perform this operation at most k times.

# Return the length of the longest substring containing the same letter 
# you can get after performing the above operations.

# The window constraint is window_lenth - max_frequency <= k
from collections import Counter
def character_replacement(s, k):
    window = Counter()
    left = 0
    max_freq = 0
    longest_substring = 0

    for right in range(len(s)):
        character = s[right]
        window[character] += 1

        max_freq = max(max_freq, window[character])

        # window condition is broken and move left side of window along
        if (right - left + 1) - max_freq > k:
            left_char = s[left]
            window[left_char] -= 1

            left += 1

        longest_substring = max(longest_substring, right - left + 1)
    
    return longest_substring
````

````python
import math

def minSubArrayLen(target, nums):
    """
        Find the minimum subarray with the given target sum.
        **Note**: Use dynamic window to solve this problem
        Consist of expanding the window and shrinking the window
    """
    if not target or not nums:
        return 0
    
    left = 0
    window_sum = 0

    min_length = math.inf

    for right in range(len(nums)):
        # expand the window until we get to the target sum
        # usually on the right side
        window_sum += nums[right]

        # shrink the window to the smallest length
        # (usually on the left side)
        while window_sum >= target:
            current_length = right - left + 1
            min_length = min(min_length, current_length)

            window_sum -= nums[left]
            left += 1
    
    return 0 if min_length == math.inf else min_length
````

````python
def length_of_longest_substring(s: str) -> int:
    """
        Find the length of the longest substring without repeating characters
    """
    # use a set to check for duplicate
    window_chars = set()

    left = 0
    max_length = 0

    for right in range(len(s)):

        # shrink until there is no duplicate
        while s[right] in window_chars:
            window_chars.remove(s[left])
            left += 1

        # expand automatically to try to get the longest substring
        window_chars.add(s[right])
        current_length = right - left + 1
        max_length = max(max_length, current_length)

    return max_length
````

### prefix and suffix product with delay

````python
# Product of Array Except Self
# Given an integer array nums, return an array answer such that answer[i] is equal to the product of all the elements of nums except nums[i].#
# The product of any prefix or suffix of nums is guaranteed to fit in a 32-bit integer.#
# You must write an algorithm that runs in O(n) time and without using the division operation.

## The key insight is by delaying the product sum by 1, we calculate all the product to the left of each indices.
## And going backward delaying by 1, we calculate all the product to the right.
def product_array_except_self(nums):
    output = [1] * len(nums)

    prefix = 1
    for i in range(len(nums)):
        output[i] = prefix
        prefix *= nums[i]
    
    postfix = 1
    for i in range(len(nums) - 1, -1, -1):
        output[i] *= postfix
        postfix *= nums[i]

    return output
````

````python
# Given an integer array nums and an integer k, return true if nums has a good subarray or false otherwise.
# A good subarray is a subarray where:
#  its length is at least two, and
#  the sum of the elements of the subarray is a multiple of k.
# Note that:
#
#  A subarray is a contiguous part of the array.
#  An integer x is a multiple of k if there exists an integer n such that x = n * k. 0 is always a multiple of k.
def continuous_array_sum(nums, k):
    remainder_map = {0: -1}
    total = 0
    
    for i in range(len(nums)):
        total += nums[i]
        if k != 0:
            total = total % k
        
        if total not in remainder_map:
            remainder_map[total] = i
        elif i - remainder_map[total] > 1:
            return True
    
    return False
````

spiral matrix
- Given a matrix, return the spiral - starting from left to right and keeping in a spiral motion

````python
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

# return  a spiral

top = 0
bottom = len(matrix) - 1
left = 0
right = len(matrix[0]) - 1
results = []
while top <= bottom and left <= right:

    # going from left to right
    for col in range(left, right + 1):
        results.append(matrix[top][col])
    top += 1

    # going from top to bottom
    for row in range(top, bottom + 1):
        results.append(matrix[row][right])
    right -= 1

    # going from right to left
    if top <= bottom:
        for col in range(right, left - 1, -1):
            results.append(matrix[bottom][col])
        bottom -= 1

    # going from bottom to top
    if left <= right:
        for row in range(bottom, top - 1, -1):
            results.append(matrix[left][left])
        left += 1
print(results)
````

### Decreasing Stack
- Use a decreasing stack

#### Next greater element
- simplest form of using a decreasing stack

````python
# Find the next greater element
# return an arra
nums = [2, 3, 4, 0, 1, 5]
stack = []
results = [-1] * len(nums)
for i, num in enumerate(nums):
    
    # if num is greater than stack, then set it for all the indices
    # that the number is greater 
    while stack and num > nums[stack[-1]]:
        position = stack.pop()
        results[position] = num
    
    # push smaller elements
    stack.append(i)
    
print(results)
````

### Reverse a Linked List
- Used 3 pointers to keep track of the previous, current, and next pointers

````python

def reverse_linked_list(root):
    prev = None
    cur = root
    
    while cur:
        nxt = cur.next
        cur.next = prev
        prev = cur
        cur = nxt
        
    return prev
````

### Two stacks

- **Implement Queue using Stacks**
- Implement a first in first out (FIFO) queue using only two stacks. The implemented queue should support all the functions of a normal queue (push, peek, pop, and empty).

- Implement the MyQueue class:
    - void push(int x) Pushes element x to the back of the queue.
    - int pop() Removes the element from the front of the queue and returns it.
    - int peek() Returns the element at the front of the queue.
    - boolean empty() Returns true if the queue is empty, false otherwise.

- **Notes**:
    - You must use only standard operations of a stack, which means only push to top, peek/pop from top, size, and is empty operations are valid.
    - Depending on your language, the stack may not be supported natively. You may simulate a stack using a list or deque (double-ended queue) as long as you use only a stack's standard operations.

````python
class MyQueue:
    def __init__(self):
        self.stack_in = []  # Used for pushing elements
        self.stack_out = []  # Used for popping/peeking elements

    def push(self, x: int) -> None:
        """Pushes element x to the back of the queue."""
        self.stack_in.append(x)

    def pop(self) -> int:
        """Removes the element from the front of the queue and returns it."""
        self._transfer_elements()
        return self.stack_out.pop()

    def peek(self) -> int:
        """Returns the element at the front of the queue."""
        self._transfer_elements()
        return self.stack_out[-1]

    def empty(self) -> bool:
        """Returns true if the queue is empty, false otherwise."""
        return not self.stack_in and not self.stack_out

    def _transfer_elements(self):
        """Helper function to move elements from stack_in to stack_out if stack_out is empty."""
        if not self.stack_out:
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())
````

````python
# LeetCode: Dota2 Senate
# This is a Dota2 Senate voting problem with two parties: Radiant ('R') and Dire ('D'). 
# Senators vote in rounds following the given string order, and 
# each can either ban an opponent's voting rights or announce victory if only their party remains. 
# Banned senators are skipped in subsequent rounds, and
# the process continues until one party eliminates all opposition. 
# Each senator plays optimally for their party by banning the next opposing senator in sequence. 
# The task is to predict which party wins given the initial senate string.
def predictPartyVictory(senate):
    n = len(senate)
    dire_queue = deque()
    radiant_queue = deque()

    for i, senator in enumerate(senate):
        if senator == "D":
            dire_queue.append(i)
        else:
            radiant_queue.append(i)

    while dire_queue and radiant_queue:
        index_d = dire_queue.popleft()
        index_r = radiant_queue.popleft()

        if index_d < index_r:
            dire_queue.append(index_d + n)
        else:
            radiant_queue.append(index_r + n)

    return "Radiant" if radiant_queue else "Dire"

# while 2 queues are optimal, 1 queue is also possible
def predictPartyVictory2(senate):
    q = deque(senate)
    radiant_bans = 0
    dire_bans = 0
    n = len(senate) 
    
    while q:
        current_senator = q.popleft()

        if current_senator == 'R':
            if dire_bans > 0:
                dire_bans -= 1
            else:
                radiant_bans += 1 
                q.append('R') 
        else: 
            if radiant_bans > 0:
                radiant_bans -= 1 
            else:
                dire_bans += 1 
                q.append('D')
        
    return "Radiant" if q[0] == 'R' else "Dire"
````

## 13. Trie
- A trie, also known as a prefix tree or digital tree, is a specialized tree-like data structure used for storing and efficiently retrieving a collection of strings or sequences of symbols. 
- It is particularly effective for operations involving prefix matching. 
- Tree-based structure: Tries are organized as a tree, where each node typically represents a character or symbol.
- Prefix-based organization: Strings are stored in a way that common prefixes share common nodes in the trie. For example, "car", "cat", and "care" would all share the nodes representing 'c' and 'a'.
- Nodes and Edges:
    - The root node represents an empty string and typically has no associated character.
    - Edges: connect parent nodes to their children, representing the transition to the next character in a string.
    - Child nodes: store the subsequent characters in a string.
- Word Markers: Nodes that represent the end of a complete word are often marked as "terminal nodes" or "word nodes" to distinguish them from nodes that are only part of a longer prefix.
- Efficient Operations: Tries allow for efficient insertion, deletion, and searching of strings. Crucially, they excel at prefix-based searches, such as finding all words starting with a particular sequence of characters.

- Common Applications:
    - Autocompletion and predictive text: Suggesting words as a user types.
    - Spell checking: Identifying misspelled words and suggesting corrections.
    - Dictionary implementation: Efficiently storing and searching a large set of words.
    - IP routing tables: Storing and searching IP addresses.
    - Bioinformatics: Analyzing DNA sequences.

- Advantages:
    - Fast prefix-based searches: Finding strings with common prefixes is highly efficient.
    - Space efficiency for common prefixes: Shared prefixes reduce redundant storage.
    - Alphabetical sorting: Strings can be retrieved in alphabetical order by traversing the trie.

- Disadvantages:
    - Memory consumption: Can consume more memory than hash tables for sparse datasets, especially if the alphabet is large.
    - Complexity for non-string data: Less suitable for data that cannot be easily represented as sequences of symbols.


````python
from collections import deque

class TrieNode:
    """A node in the Trie data structure."""
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False

class Trie:
    """A Prefix Trie implementation with insert, search, and prefix search capabilities."""
    def __init__(self):
        self.root = TrieNode()
        self.word_count = 0
        
    def build(self, words):
        """Build the trie from a list of words"""
        for word in words:
            self.insert(word)
    
    def insert(self, word):
        """Inserts a word into the trie."""
        node = self.root
        
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        
        if not node.is_end_of_word:
            node.is_end_of_word = True
            self.word_count += 1
            
    
    def search(self, prefix):
        """Checks if a complete word exists in the trie."""
        node = self.root
        
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        
        return node.is_end_of_word
    
    def starts_with(self, prefix):
        """Checks if there is any word in the trie that starts with the given prefix."""
        node = self.root
        
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        
        return True
    
    def collect_words(self, node, prefix):
        
        words = []
        
        if node.is_end_of_word:
            words.append(prefix)
        
        for char, child_node in node.children.items():
            words.extend(self.collect_words(child_node, prefix + char))
            
        return words

    def delete(self, word):
        """Deletes a word from the trie, backtracking to remove unnecessary nodes."""

        def _delete_recursive(node, word, index):
            if index == len(word):
                # Reached the end of the word path
                if not node.is_end_of_word:
                    return False  # Word doesn't exist to delete
                node.is_end_of_word = False
                # Return True if this node can be deleted (no other words share this prefix/node)
                return len(node.children) == 0

            char = word[index]
            if char not in node.children:
                return False  # Character path doesn't exist

            # Recursively call delete on the child node
            can_delete_child = _delete_recursive(
                node.children[char], word, index + 1)

            if can_delete_child:
                # If the child can be deleted, remove the link from the current node
                del node.children[char]
                # Return True if the current node should also be deleted (no children left AND not end of another word)
                return len(node.children) == 0 and not node.is_end_of_word

            return False  # Cannot delete current node

        if _delete_recursive(self.root, word, 0):
            self._word_count -= 1
            return True
        return False
    
    
    
    def get_suggestions(self, prefix):
        node = self.root
        suggestions = []
        
        for char in prefix:
            if char not in node.children:
                return []

            node = node.children[char]
            
        def dfs_collect(node, current_word_chars):
            if node.is_end_of_word:
                suggestions.append("".join(current_word_chars))
            
            for char, child_node in node.children.items():
                current_word_chars.append(char)
                dfs_collect(child_node, current_word_chars)
                current_word_chars.pop()
                
        dfs_collect(node, list(prefix))
        return suggestions
        
    def get_longest_common_prefix(self):
        node = self.root
        lcp = []
        
        while len(node.children) == 1 and not node.is_end_of_word:
            char = next(iter(node.children))
            node = node.children[char]
            lcp.append(char)
        
        return "".join(lcp)
    
    def get_shortest_word(self):
        queue = deque([(self.root, "")])
        
        while queue:
            node, word = queue.popleft()
            
            if node.is_end_of_word:
                return word
            
            for char, child_node in node.children.items():
                queue.append((child_node, word + char))
        
        return None
````

## 14. Heap or Priority Queues
- Structure: A heap is typically implemented as a complete binary tree, meaning all levels are fully filled, except possibly the last level, which is filled from left to right. This structure allows efficient array representation.
- Heap Property: Heaps maintain a specific ordering among parent and child nodes. There are two main types:
    - Max-Heap: The value of a parent node is always greater than or equal to the values of its children. The largest element is always at the root.
    - Min-Heap: The value of a parent node is always less than or equal to the values of its children. The smallest element is always at the root.

- Implementation: Due to its complete binary tree nature, a heap is most often implemented using a simple array or list, where the relationships between parent and child nodes can be calculated using array indices.

- Peak Element Access: The element with the highest (Max-Heap) or lowest (Min-Heap) priority is always at the root of the tree (index 0 in an array implementation), allowing for $O(1)$ access (constant time).

- Insertion (Push) Time Complexity: Inserting a new element into a heap takes $O(\log n)$ time, where $n$ is the number of elements. This involves placing the new element at the end and then performing a "heapify-up" or "sift-up" operation to restore the heap property.Deletion (Pop/Extract) Time Complexity: 

- Deleting the root (the max/min element) also takes $O(\log n)$ time. This involves replacing the root with the last element, removing the last element, and then performing a "heapify-down" or "sift-down" operation to restore the heap property.

- Primary Application: Heaps are the fundamental data structure used to efficiently implement a Priority Queue, and they are also the basis for the Heapsort sorting algorithm.

````python
import heapq


class MinHeap:
    def __init__(self):
        # The heap is stored as a simple list
        # no need to call headq.heapify because list is empty
        self.heap = []

    def push(self, item):
        """Adds an item to the heap."""
        # heapq.heappush maintains the min-heap property
        heapq.heappush(self.heap, item)

    def pop(self):
        """Removes and returns the smallest item from the heap."""
        # heapq.heappop removes and returns the smallest item (root)
        if self.heap:
            return heapq.heappop(self.heap)
        return None

    def peek(self):
        """Returns the smallest item without removing it."""
        # The smallest item is always at index 0
        if self.heap:
            return self.heap[0]
        return None

    def is_empty(self):
        """Checks if the heap is empty."""
        return len(self.heap) == 0
````

````python

class MaxHeap:
    def __init__(self):
        # The heap is stored as a simple list
        self.heap = []

    def push(self, item):
        """Adds an item to the heap as its negative value."""
        # Store the negative value to simulate max-heap behavior
        heapq.heappush(self.heap, -item)

    def pop(self):
        """Removes and returns the largest item (by negating the result)."""
        if self.heap:
            # Negate the popped value to get the original maximum
            return -heapq.heappop(self.heap)
        return None

    def peek(self):
        """Returns the largest item without removing it."""
        if self.heap:
            # Negate the value at index 0 to get the original maximum
            return -self.heap[0]
        return None

    def is_empty(self):
        """Checks if the heap is empty."""
        return len(self.heap) == 0
````

#### Heap from scratch
- Below is the implementation for a heap from scratch without external library

````python
class PriorityQueue:
    def __init__(self):
        self.heap = []

    def _swap(self, left, right):
        self.heap[left], self.heap[right] = self.heap[right], self.heap[left]
    
    def _parent(self, index):
        return (index - 1) // 2
    
    def _left(self, parent):
        return parent * 2 + 1
    
    def _right(self, parent):
        return parent * 2 + 2

    def _heapify_up(self, index):
        """
        Maintains the heap property after an insertion (bottom up).
        Compare child with parent and swaps if child priority is higher.
        """
        parent_index = self._parent(index)

        while index > 0 and self.heap[index] < self.heap[parent_index]:
            self._swap(index, parent_index)
            index = parent_index
            parent_index = self._parent(index)

    def _heapify_down(self, index):
        """
        Maintains the heap property after an extraction (top down).
        Compareprent with child and swaps with smallest child
        """
        left = self._left(index)
        right = self._right(index)
        smallest = index 
        heap_size = len(self.heap)

        if left < heap_size and self.heap[left] < self.heap[smallest]:
            smallest = left

        if right < heap_size and  self.heap[right] < self.heap[smallest]:
            smallest = right
        
        if smallest != index:
            self._swap(index, smallest)
            index = smallest
        
    def push(self, item):
        self.heap.append(item)
        self._heapify_up(len(self.heap) - 1)

    def pop(self):
        self._swap(0, len(self.heap) -1)
        item = self.heap.pop()
        self._heapify_down(0)
        return item

    def peek(self):
        return self.heap[0] if self.heap else None 


    def __len__(self):
        return len(self.heap)
````

### Sorting by end times
- Sorting by end time is the "golden rule" for interval problems because it allows you to process options in the order they become unavailable.

### Merge Intervals
- Given an array of intervals where intervals[i] = [starti, endi], merge all overlapping intervals, and return an array of the non-overlapping intervals that cover all the intervals in the input.

````python
# Merge overlapping intervals with a single pass
def merge(intervals):
    # sort by the start
    intervals.sort(key = lambda i: i[0])
    results = [intervals[0]] # add the smallest interval first

    # compare the current interval start time with the end time of the last interval
    # in the result list
    for start, end in intervals[1:]:
        last_end = results[-1][1]
        if start <= last_end:
            results[1][1] = max(last_end, end)
        else:
            results.append([start, end])
    return results 
````

### Overlapping Intervals/Intersections
- You are given two lists of closed intervals, firstList and secondList, where firstList[i] = [starti, endi] and secondList[j] = [startj, endj]. Each list of intervals is pairwise disjoint and in sorted order.

- Return the intersection of these two interval lists.

- A closed interval [a, b] (with a <= b) denotes the set of real numbers x with a <= x <= b.

- The intersection of two closed intervals is a set of real numbers that are either empty or represented as a closed interval. For example, the intersection of [1, 3] and [2, 4] is [2, 3].

````python
def intervalInsection(first_list, second_list):
    results = []
    i = 0
    j = 0
    
    while i < len(first_list) and j < len(second_list):
            list_first = first_list[i]
            list_second = second_list[i]

            low = max(list_first[0], list_second[0])
            high = min(list_first[1], list_second[1])

            if low <= high:
                results.append([low, high])

            if list_first[1] < list_second[1]:
                i += 1
            else:
                j += 1
        
    return  results
````

### Sweep line  algorithm
- a powerful pattern used to solve various geometric and interval problems, often when simple sorting or two-pointers isn't enough, especially for finding the maximum number of simultaneous overlaps or complex intersections.
- The core idea is to transform the 2D intervals into a list of 1D events (start and end points) and process these events chronologically to track the state of the system.

````python
# Finding the maximum number of concurrent meetings/intervals
def max_concurrent_intervals(intervals):
    # 1. Create Event Points (The "Sweep Line")
    # Events are stored as tuples: (time, type).
    # Type: +1 for a start point (event starts), -1 for an end point (event ends).
    # Note: We prioritize -1 (end) over +1 (start) for identical times 
    # to correctly count overlaps at endpoints.
    events = []
    for start, end in intervals:
        events.append((start, 1))
        events.append((end, -1))

    # 2. Sort the Events
    # Sort primarily by time. 
    # Secondary sort: Ensures that if events happen at the same time, 
    # the END event (-1) is processed before the START event (+1). 
    # This prevents the start of a new interval from being counted 
    # *before* # an old one ends at the same time, which is crucial for strict "overlap."
    events.sort()

    max_overlap = 0
    current_overlap = 0

    # 3. Sweep the Line and Track the Count
    # Update the current number of concurrent intervals
    # Update the maximum overlap found so far
    for time, type in events:
        current_overlap += type

        max_overlap = max(max_overlap, current_overlap)
    
    return max_overlap
````

### Dijkstra's Algorithm
- Dijkstra's Algorithm: A famous shortest path algorithm that finds the shortest paths from a single starting vertex to all other vertices in a weighted graph with **non-negative edge weights**. It is widely used in network routing.
- Runtime:  (O(|E|+|V|\log |V|)) , V is vertices and E is number of edges

````python
import heapq
# generic Dijkstra's Algorithm
def dijkstra(graph, start):
    distances = {node: float("inf") for node in graph}
    distances[start] = 0
    priority_queue = [(0, start)]
    
    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)
        
        if current_distance > distances[current_node]:
            continue 
        
        for neighbor, weights in graph[current_node].items():
            distance = current_distance + weights 
            
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                # eagerly push the better distance into the queue immediately
                heapq.heappush(priority_queue, (distance, neighbor))
                
    return distances
        
````

### Bellman-Ford Algorithm
- Another shortest path algorithm, similar to Dijkstra's, but capable of handling graphs that contain negative edge weights. It can also detect if a graph has a negative-weight cycle.
- Use dijkstra for non-negative values
- The relaxation step helps detect negative cycles in Bellman-Ford because the algorithm inherently relies on the principle that the shortest path in a graph with \(|V|\) vertices can have at most \(|V|-1\) edges.
- The main loop of the Bellman-Ford algorithm iterates exactly \(|V|-1\) times. The goal of these iterations is to ensure that all shortest paths are found, as the longest possible simple path in the graph (a path that doesn't revisit any vertex) can traverse at most all other \(|V|-1\) vertices.
- After \(|V|-1\) iterations, every distance distances[v] holds the true shortest distance from the source unless there is a negative cycle reachable from the source.
- The detection mechanism happens after the main loop has completed its \(|V|-1\) passes. The algorithm performs a final, \(|V|\)-th iteration over all edges:
    - If all distances remain the same during this final pass, it confirms that the distances found in the previous steps are final and optimal. 
    - No negative cycle exists.If a distance can still be reduced during this extra iteration, it means a path is being found that is even shorter than the one previously marked as the "shortest path".
- A path that continues to get shorter with every iteration, even after the theoretical maximum number of edges has been considered, can only happen if you are traversing a loop that has a cumulative negative weight.
- If a negative cycle exists, you can theoretically loop through it infinitely many times, making the overall path weight approach negative infinity. The relaxation step would infinitely reduce the distance value.
- The Bellman-Ford algorithm stops after the \(|V|\)-th check and declares a negative cycle the moment it finds that a path can still be "relaxed" or made shorter

- Runtime: O(|V|  |E|), V number of vertices, and E is number of edges


````python
# generic algorithm for Bellman-Ford Algorithm

def bellman_ford(vertices, edges, source):
    """
    Finds the shortest paths from a source node using the Bellman-Ford algorithm,
    and detects negative cycles.
    Args:
        vertices (list): A list of all vertex IDs in the graph.
        edges (list): A list of tuples (source, destination, weight).
        source: The starting node ID.

    Returns:
        dict: A dictionary of shortest distances, or a string indicating 
              a negative cycle was found.
    """
    distances = {v: float("inf") for v in vertices }
    distances[source] = 0
    
    # Relax edges v -1 times
    for _ in range(len(vertices) - 1):
        for u, v, weight in edges:
            if distances[u] != float("inf") and distances[u] + weight < distances[v]:
                distances[v] = distances[u] + weight
    
    # Check for negative weight cycles
    for u, v, weight in edges:
        if distances[u] != float("inf") and distances[u] + weight < distances[v]:
            return None # there is a negative weight cycles
    
    return distances
    
````

### Union Find
- A method to connect graphs together into 1 big graph

````python
n = 50
parent = list(range(n + 1))
"""
find parent and returns it (recursion)
"""


def find(i):
    if parent[i] == i:
        return i
    parent[i] = find(parent[i])
    return parent[i]


"""
Union (naive approach)
"""


def union(i, j):
    root_i = find(i)
    root_j = find(j)
    if root_i != root_j:
        parent[root_i] = root_j
        return True
    return False
````

````python
# the union find algorithm inside a class with ranking
class DisjointSetUnion:
    
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n 
        
    def find(self, i):
        if self.parent[i] == i:
            return i
        self.parent[i] = self.find(self.parent[i])
        
        return self.parent[i]
    
    def union(self, i, j):
        root_i = self.find(i)
        root_j = self.find(j)
        if root_i != root_j:
            if self.rank[root_i] < self.rank[root_j]:
                self.parent[root_i] = root_j 
            elif self.rank[root_i] > self.rank[root_j]:
                self.parent[root_j] = root_i
            else:
                self.parent[root_j] = root_i 
                self.rank[root_i] += 1
            return True 
        return False 


    
````

### Kruskal's Algorithm (MST) Minimum Spanning Tree
- A greedy algorithm used to find the Minimum Spanning Tree (MST) for a connected, weighted, undirected graph. An MST is a subset of the edges that connects all the vertices together, without any cycles, and with the minimum possible total edge weight.
- The algorithm works by iteratively adding the cheapest edge that connects two previously disconnected components, without forming a cycle. 
- Required a Union Find data structure or algorithm to work

````python

# 
def kruskal(vertices, edges):
    """
    Finds the Minimum Spanning Tree (MST) of a graph using Kruskal's algorithm.

    Args:
        vertices: The number of vertices in the graph.
        edges: A list of tuples, where each tuple represents an edge (u, v, weight).

    Returns:
        A list of edges forming the MST, or an empty list if no MST can be formed.
    """
    mst = []
    dsu = DisjointSetUnion(vertices)
    
    # Sort all edges by weight in ascending order
    edges.sort(key=lambda item: item[2])
    
    num_edges_in_mst = 0
    for u, v, weight in edges:
        if dsu.union(u, v):
            mst.append((u, v, weight))
            num_edges_in_mst += 1
            if num_edges_in_mst == vertices - 1:
                break 
    return mst 
````

### Prim's Algorithm (MST) Minimum Spanning Tree
- An alternative greedy algorithm, also used to find the Minimum Spanning Tree (MST). Unlike Kruskal's, which builds a forest of trees, Prim's algorithm builds a single tree by continually adding the cheapest edge from a vertex already in the tree to one not yet in the tree.
- DOES NOT NEED A UNION-FIND Algorithm, BUT a **PRIORITY QUEUE**

````python
import heapq

def prim_mst(graph, start_node):
    """
    Implements Prim's Algorithm to find the Minimum Spanning Tree (MST).

    Args:
        graph (dict): An adjacency list where graph[u] = [(weight, v), ...].
        start_node (int): The node to begin the algorithm from.

    Returns:
        list: A list of edges forming the MST, format: [(u, v, weight), ...] and total weight.
    """
    
    min_heap = [(0, -1, start_node)]  # Weight, Parent Node, Current Node
    visited = set()
    mst = []
    total_weights = 0
    
    while min_heap:
        weight, parent_node, u = heapq.heappop(min_heap)
        
        if u in visited:
            continue 
        
        visited.add(u)
        total_weights += weight
        
        if parent_node != -1:
            mst.append((parent_node, u, weight))
        
        for neighbor_weight, v in graph.get(u, []):
            if v not in visited:
                heapq.heappush(min_heap, (neighbor_weight, u, v))
                
    return mst, total_weights
````

### Floyd-Warshall Algorithm: 
- An algorithm used to find the shortest paths between all pairs of vertices in a weighted graph. It is an example of a dynamic programming approach.
- After the main loops, the algorithm checks for negative weight cycles. If dist[i][i] (the shortest path from a vertex to itself) is negative for any i, it indicates the presence of a negative weight cycle. In such cases, shortest paths are not well-defined, and the function prints a message and returns None.
- Best for relatively small, dense networks where you need the shortest distance between every single location within the network, rather than just one specific starting point


- Scenario: A Corporate Campus Intranet
    - Imagine a large office campus with several buildings, each housing a main network switch (representing the vertices in the graph). The cables connecting these switches have different capacities and latencies (representing edge weights). 
    - The network administrator needs every router to efficiently determine the absolute best path to reach every other building's switch dynamically. 

    - How Floyd-Warshall is applied:
        - Graph Representation: The network topology is modeled as a weighted, directed graph where:
            - Vertices (V) are the switches/buildings (e.g., V = 20 buildings).
            - Weights are metrics like data latency or hop count between switches.
    - The Goal: The algorithm calculates the complete "all-pairs shortest paths" matrix.
    - The Result: Each router (vertex) receives a lookup table that instantly tells it the shortest path (lowest latency) to forward data to any other destination building on the campus network.
    - Efficiency: Because the number of buildings is relatively small (e.g., 20-50), the \(O(V^{3})\) complexity of Floyd-Warshall is efficient enough for the system to run the update whenever the network topology changes (e.g., a cable is upgraded or fails).
- In summary, it shines when you need a comprehensive, pre-calculated "mileage chart" for every possible origin and destination pair within a system.

- Time: **O(V) (Big O of V cubed)**
- Space: **O(V) (Big O of V squared)**


````python


def floyd_warshall(graph):
    """
    Finds the shortest paths between all pairs of vertices in a weighted graph
    using the Floyd-Warshall algorithm.

    Args:
        graph: A 2D list (matrix) representing the adjacency matrix of the graph.
               graph[i][j] is the weight of the edge from vertex i to vertex j.
               Use float('inf') to represent no direct edge.
               Diagonal elements graph[i][i] should be 0.

    Returns:
        A 2D list (matrix) where result[i][j] is the shortest path length
        from vertex i to vertex j.
    """
    num_vertices = len(graph)
    
    dist = [row[:] for row in graph]
    for k in range(num_vertices):
        for i in range(num_vertices):
            for j in range(num_vertices):
                if dist[i][k] + dist[k][j] < dist[i][j]:  
                    dist[i][j] = dist[i][k] + dist[k][j]
                    
    for i in range(num_vertices):
        if dist[i][i] < 0:
            return None 
    
    return dist
````

### Topological Sort
- Kahn's algorithm is an approach to topological sorting that utilizes a Breadth-First Search (BFS) like strategy. 
- It works by iteratively identifying and processing nodes with an in-degree of zero.

````python
from collections import deque 

def kahn_topological_sort(num_nodes, edges):
    """
    Performs topological sorting using Kahn's algorithm.

    Args:
        num_nodes: The total number of nodes in the graph.
        edges: A list of tuples representing directed edges (u, v)
               where u -> v.

    Returns:
        A list representing a topological ordering of the graph,
        or an empty list if a cycle is detected.
    """
    adj = [[] for _ in range(num_nodes)]
    in_degree = [0] * num_nodes
    
    for u, v in edges:
        adj[u].append(v)
        in_degree[v] += 1
    
    queue = deque()
    for i in range(num_nodes):
        if in_degree[i] == 0:
            queue.append(i)
        
    
    topological_order = []
    nodes_visited = 0
    
    while queue:
        u = queue.popleft()
        topological_order.append(u)
        nodes_visited += 1
        
        for v in adj[u]:
            in_degree[v] -= 1
            if in_degree[v] == 0:
                queue.append(v)
    
    
    return topological_order if nodes_visited == num_nodes else []
    
````

### Traveling Salesman Problem (Held-Karp algorithm) Dynamic Programming
- finding the shortest possible route that visits every node (city) exactly once and returns to the starting node. 
- This is a classic NP-hard problem, meaning optimal solutions become computationally expensive for large numbers of nodes.
- Due to its NP-hard nature, finding the absolute optimal solution for a large number of cities requires enormous computational power. 
- The implementations usually rely on heuristics or approximation algorithms for efficient results
- Time complexity is **(O(n^{2}\times 2^{n}))**


````python
import sys
# Use dynamic programming with bit manipulation
# the total number of cities should be less than 20ish otherwise it is too large for the algorithm
def solve_tsp(distances):
    n = len(distances)
    memo = {}
    
    start_node = 0
    full_mask = (1 << n) - 1
    
    def held_karp(mask, current_city):
        if mask == full_mask:
            return distances[current_city][start_node]
        
        if (mask, current_city) in memo:
            return memo[(mask, current_city)]
        
        min_distance = sys.maxsize
        
        for next_city in range(n):
            if not (mask & (1 << next_city)):
                new_mask = mask | (1 <<next_city)
                distance = distances[current_city][next_city] + held_karp(new_mask, next_city)
                if distance < min_distance:
                    min_distance = distance 
                    
        memo[(mask, current_city)] = min_distance 
        return min_distance 
    
    initial_mask = (1 << start_node)
    min_tour_cost = held_karp(initial_mask, start_node)
    return min_tour_cost
    
````

````python
distances = [
    [0, 10, 15, 20],  # Distances from A to A, B, C, D
    [10, 0, 35, 25],  # Distances from B to A, B, C, D
    [15, 35, 0, 30],  # Distances from C to A, B, C, D
    [20, 25, 30, 0]   # Distances from D to A, B, C, D
]
min_cost = solve_tsp(distances)
print(f"The minimum tour cost is: {min_cost}")
# Another example:
distances_2 = [
    [0, 2, 9, 10],
    [1, 0, 6, 4],
    [15, 7, 0, 8],
    [6, 3, 12, 0]
]

min_cost2 = solve_tsp(distances_2)
print(f"The minimum tour cost is: {min_cost2}")
````

### Eulerian Paths
- a trail in a finite graph that visits every edge exactly once (allowing for revisiting vertices). 
- The graph must be connected, and have exactly zero or two vertices with an odd degree (number of edges).
- Two main algorithm: Fleury and Hierholzer
- Example: 
    - **Street Sweeping/Mail Delivery**: A city planner might use an Eulerian path to find the most efficient route for a street sweeper to cover every street (edge) exactly once

### Fleury's algorithm
- An algorithm for finding eulerian path
- Builds the path by traversing edges one at a time, avoiding bridges (edges that, if removed, disconnect the remaining graph) unless it is the only option.
- Requires repeated checks for bridges in the remaining graph at each step to maintain connectivity.
- Time: O(E^2):  due to the repeated need to check for bridges.
- Less efficient, better suited for small to medium-sized graphs or for educational purposes.

````python
from collections import defaultdict, deque
import copy


class Graph:
    def __init__(self, vertices):
        self.V = vertices
        # Adjacency list stored as a dictionary of lists
        self.graph = defaultdict(list)
        self.path = []

    def add_edge(self, u, v):
        """Add an undirected edge to the graph."""
        self.graph[u].append(v)
        self.graph[v].append(u)

    def remove_edge(self, u, v):
        """Remove an edge from the graph."""
        self.graph[u].remove(v)
        self.graph[v].remove(u)

    def dfs_count(self, u, visited):
        """DFS to count reachable vertices from u."""
        visited[u] = True
        count = 1
        for v in self.graph[u]:
            if not visited[v]:
                count += self.dfs_count(v, visited)
        return count

    def is_bridge(self, u, v):
        """Check if edge (u, v) is a bridge."""
        # Count vertices reachable from u before removing the edge
        visited1 = {node: False for node in self.graph}
        count1 = self.dfs_count(u, visited1)

        # Remove the edge (u, v) temporarily
        self.remove_edge(u, v)

        # Count vertices reachable from u after removal
        visited2 = {node: False for node in self.graph}
        count2 = self.dfs_count(u, visited2)

        # Add the edge back
        self.add_edge(u, v)

        # If count1 > count2, removing the edge increased disconnected components
        return count1 > count2

    def find_eulerian_path(self, start_vertex):
        """Applies Fleury's algorithm to find the Eulerian path/circuit."""
        current_vertex = start_vertex
        while any(self.graph.values()):  # While there are edges left
            # Make a copy to iterate
            neighbors = list(self.graph[current_vertex])
            for next_vertex in neighbors:
                # Rule: Don't choose a bridge unless it's the only edge left
                if len(self.graph[current_vertex]) == 1 or not self.is_bridge(current_vertex, next_vertex):
                    # Add edge to path and remove from graph
                    self.path.append((current_vertex, next_vertex))
                    self.remove_edge(current_vertex, next_vertex)
                    current_vertex = next_vertex
                    break
        return self.path
````

### Hierholzer's algorithm
- Finds disjoint cycles using a modified Depth First Search (DFS) and then merges them to form the final path/circuit.
- Does not require explicit bridge checks.
- Time Complexity: (O(V+E)) (linear time), as it visits each vertex and edge essentially once.
- Highly efficient, making it the preferred method for large and complex graphs in practice.

````python
from collections import defaultdict


def hierholzer(graph_adj):
    """
    Finds an Eulerian circuit in a directed graph using Hierholzer's algorithm.

    Args:
        graph_adj: A dictionary representing the graph as an adjacency list, 
                   e.g., {0: [1, 2], 1: [2], 2: [0]}.

    Returns:
        A list of vertices representing the Eulerian circuit.
    """

    # Check for empty graph
    if not graph_adj:
        return []

    # Find a starting node that has at least one edge
    start_node = next(iter(graph_adj.keys()))

    # Stacks for path management
    current_path = [start_node]
    circuit = []

    # The algorithm runs as long as the current path has vertices
    while current_path:
        current_node = current_path[-1]

        # If the current node still has untraversed edges
        if graph_adj[current_node]:
            # Pick the next adjacent vertex and remove the edge immediately
            next_node = graph_adj[current_node].pop()
            current_path.append(next_node)
        else:
            # If no more outgoing edges from current_node,
            # this part of the circuit is complete.
            # Pop the vertex from current_path and add to the final circuit.
            circuit.append(current_path.pop())

    # The circuit is built in reverse order, so reverse it at the end.
    return circuit[::-1]
````

### Bridges
- a bridge (or cut-edge) is an **edge** in a graph whose removal increases the number of connected components. Think of it as that one friend who always brings the group togetherremove them, and suddenly everyones awkwardly standing around, unsure of what to do.
- A bridge is an edge and an articulation is a vertex or node.
- Its removal increases the number of connected components in the graph.
- Every bridge connects two components when removed.

### Articulations
-  An articulation is a **vertex** whose removal, along with all its incident edges, increases the number of connected components in the graph. 
- an articulation point is a critical node that maintains the connectivity of separate parts of the graph. If you think of a network of roads or a communication grid, an articulation point represents a single point of failure: if that specific location fails, parts of the network become isolated.
- Its removal (and all incident edges) increases the number of connected components.
- An articulation point can be an endpoint of a bridge, but not necessarily all articulation points are endpoints of bridges, and not all edges connected to an articulation point are bridges.

- **Relationship Between the Two**
    - If a graph has a bridge, at least one of its endpoints is an articulation point, unless that endpoint has a degree of 1 (a "leaf" node).
    - It is possible for a graph to have articulation points without having any bridges. For example, in a graph made of two triangles connected at a single vertex, the connecting vertex is an articulation point, but none of the edges within the triangles are bridges (as they are part of a cycle). 


### Tarjan's Bridge Finding Algorithm
- The core idea is that an edge (u, v) (where u is the parent of v in the DFS tree) is a bridge if and only if no back-edge allows any node in the subtree rooted at v to reach an ancestor of u (or u itself) that was discovered earlier than u. 
- **Time Complexity: O(V + E)**, where V is the number of vertices and E is the number of edges. Yes, its efficientlike a well-oiled machine!
- **Space Complexity: O(V)** for storing visited nodes and other auxiliary data.
- **Network Design**: Ensuring that critical connections are maintained in communication networks.
- **Social Networks**: Identifying key individuals whose absence would fragment the network.
- **Infrastructure**: Analyzing road networks to find critical routes that, if closed, would isolate areas.
- **Biological Networks**: Understanding the connectivity of biological systems, like neural networks.
- **Game Development**: Designing levels where certain paths are crucial for player progression.
- **Data Analysis**: Analyzing relationships in datasets to find critical connections.
- **Graph Theory Research**: Contributing to the field of graph theory by exploring new properties of bridges.
- **Robotics**: Pathfinding algorithms in robotics often utilize bridge-finding techniques.
- **Urban Planning**: Identifying essential routes in city planning to ensure connectivity.
- **Computer Networks**: Enhancing the reliability of data transmission by identifying critical links.

````python
def tarjan(graph):
    def dfs(u, discovery, low, parent, bridges):
        nonlocal time
        discovery[u] = low[u] = time
        time += 1

        for v in graph[u]:
            if discovery[v] == -1:  # If v is not visited
                parent[v] = u
                dfs(v, discovery, low, parent, bridges)

                # Check if the subtree rooted at v has a connection back to one of the ancestors of u
                low[u] = min(low[u], low[v])

                # If the lowest vertex reachable from subtree under v is below u in DFS tree, then u-v is a bridge
                if low[v] > discovery[u]:
                    bridges.append((u, v))

            elif v != parent[u]:  # Update low value of u for parent function calls
                low[u] = min(low[u], discovery[v])

    V = len(graph)
    discovery = [-1] * V
    low = [-1] * V
    parent = [-1] * V
    bridges = []
    time = 0

    for i in range(V):
        if discovery[i] == -1:
            dfs(i, discovery, low, parent, bridges)

    return bridges
````

### Network flow
- Models systems like traffic or fluids using a directed graph with edges having capacities (max flow), a source (start), and a sink (end). 
- Flow must respect edge capacities and conserve flow at nodes (inflow = outflow, except source/sink). Key problems involve finding the maximum flow (e.g., using Ford-Fulkerson, Edmonds-Karp) or minimum cost flow, with the Max-Flow Min-Cut Theorem stating max flow equals min cut capacity, helping solve real-world problems like scheduling and matching. 
- Flow Network: A directed graph (nodes/vertices, edges/arcs) with numbers (capacities) on edges, a designated source (s), and a sink (t).
- Flow: A value assigned to each edge, representing the amount moving through it (e.g., cars, data, water).
- Capacity: The maximum flow an edge can handle.
- Capacity Constraint: Flow on an edge cannot exceed its capacity (Flow  Capacity).
- Conservation Constraint: For all nodes except s and t, total inflow equals total outflow.
- Augmenting Path: A path from source to sink in the residual graph (showing remaining capacity) that allows increasing total flow. 

- Application:
    - Logistics: Routing goods in transportation networks.
    - Computer Networks: Optimizing data traffic.
    - Matching: Assigning workers to jobs (Bipartite Matching).
    - Image Processing: Image segmentation (finding foreground/background).

### Max Flow Ford Fulkerson 
- using BFS, which guarantees finding the maximum flow in graphs with integer capacities.

````python
from collections import deque

class Graph:
    def __init__(self, graph):
        self.graph = graph  # Residual graph, stores current capacities
        self.ROW = len(graph)

    # BFS to find an augmenting path from source 's' to sink 't'
    # Fills parent[] to reconstruct the path
    def BFS(self, s, t, parent):
        visited = [False] * self.ROW
        queue = deque()

        queue.append(s)
        visited[s] = True

        while queue:
            u = queue.popleft()

            for v in range(self.ROW):
                # If v is not visited and there's a residual capacity from u to v
                if not visited[v] and self.graph[u][v] > 0:
                    queue.append(v)
                    visited[v] = True
                    parent[v] = u
                    if v == t:  # Path found to sink
                        return True
        return False

    def ford_fulkerson(self, source, sink):
        parent = [-1] * self.ROW  # Stores path
        max_flow = 0

        # While there is an augmenting path from source to sink
        while self.BFS(source, sink, parent):
            path_flow = float("Inf")
            s = sink

            # Find the minimum residual capacity (bottleneck) in the found path
            while s != source:
                path_flow = min(path_flow, self.graph[parent[s]][s])
                s = parent[s]

            # Add path flow to overall flow
            max_flow += path_flow

            # Update residual capacities of the edges and reverse edges along the path
            v = sink
            while v != source:
                u = parent[v]
                self.graph[u][v] -= path_flow  # Reduce forward edge capacity
                self.graph[v][u] += path_flow  # Increase reverse edge capacity
                v = parent[v]
        return max_flow
````

### Hopcroft karp algorithm - Unweighted Bipartite Matching 
- Finds the maximum cardinality matching in a bipartite graph. 
- It works in phases using a combination of Breadth-First Search (BFS) to find the shortest augmenting paths and Depth-First Search (DFS) to maximize the number of vertex-disjoint paths in each phase. 
- This approach gives it a superior time complexity of (O(E\sqrt{V})) compared to simpler augmenting path methods.

````python
from collections import deque

INF = 10**9


def hopcroft_karp(adj, n_left, n_right):
    """
    adj: adjacency list where adj[u] = list of right nodes v connected to left node u
         left nodes are 1..n_left
         right nodes are 1..n_right
    returns: size of maximum matching, pair_left[], pair_right[]
    """

    # pair_left[u] = matched vertex for left u
    # pair_right[v] = matched vertex for right v
    pair_left = [0] * (n_left + 1)
    pair_right = [0] * (n_right + 1)

    # distance labels for BFS
    dist = [0] * (n_left + 1)

    def bfs():
        """Builds BFS layers, returns True if an augmenting path exists."""
        q = deque()
        for u in range(1, n_left + 1):
            if pair_left[u] == 0:   # free node
                dist[u] = 0
                q.append(u)
            else:
                dist[u] = INF

        found_augmenting_path = False

        while q:
            u = q.popleft()
            for v in adj[u]:
                matched_u = pair_right[v]
                if matched_u == 0:
                    # We found a free node on the right  augmenting path exists
                    found_augmenting_path = True
                else:
                    if dist[matched_u] == INF:
                        dist[matched_u] = dist[u] + 1
                        q.append(matched_u)

        return found_augmenting_path

    def dfs(u):
        """DFS to find augmenting paths using BFS layers."""
        for v in adj[u]:
            matched_u = pair_right[v]
            if matched_u == 0 or (dist[matched_u] == dist[u] + 1 and dfs(matched_u)):
                pair_left[u] = v
                pair_right[v] = u
                return True

        dist[u] = INF
        return False

    matching = 0
    while bfs():
        for u in range(1, n_left + 1):
            if pair_left[u] == 0:
                if dfs(u):
                    matching += 1

    return matching, pair_left, pair_right
````

### Edmonds Karp Algorithm 
- the BFS-based implementation of FordFulkerson

````python
from collections import deque


def edmonds_karp(capacity, adj, source, sink):
    """
    capacity: 2D matrix capacity[u][v] = capacity of edge uv
    adj: adjacency list, adj[u] = list of neighbors v
    source: source node
    sink: sink node
    returns: maximum flow value
    """

    n = len(adj)
    parent = [-1] * n
    max_flow = 0

    def bfs():
        """Find augmenting path using BFS, return True if exists."""
        for i in range(n):
            parent[i] = -1
        parent[source] = -2   # mark source as visited

        q = deque()
        q.append((source, float('inf')))

        while q:
            u, flow = q.popleft()

            for v in adj[u]:
                if parent[v] == -1 and capacity[u][v] > 0:
                    parent[v] = u
                    new_flow = min(flow, capacity[u][v])
                    if v == sink:
                        return new_flow
                    q.append((v, new_flow))

        return 0

    while True:
        flow = bfs()
        if flow == 0:
            break    # no more augmenting paths
        max_flow += flow

        # update residual capacities
        v = sink
        while v != source:
            u = parent[v]
            capacity[u][v] -= flow
            capacity[v][u] += flow
            v = u

    return max_flow
````

````python
# Number of nodes
n = 5

capacity = [[0]*n for _ in range(n)]
adj = [[] for _ in range(n)]


def add_edge(u, v, cap):
    capacity[u][v] = cap
    adj[u].append(v)
    adj[v].append(u)  # Add reverse edge for residual graph


add_edge(0, 1, 10)
add_edge(0, 2, 5)
add_edge(1, 2, 15)
add_edge(1, 3, 10)
add_edge(2, 4, 10)
add_edge(3, 4, 10)

max_flow = edmonds_karp(capacity, adj, source=0, sink=4)
print("Max Flow =", max_flow)
````

### Dinic's Algorithm 
- A flow is maximum if there is no s to t path in residual graph.
- BFS is used in a loop. There is a difference though in the way we use BFS in both algorithms.
- O(EV)O(Emin(V^(2/3), E)) performance depending on graph type (Much faster than EdmondsKarp)

````python
from collections import deque


class Dinic:
    class Edge:
        __slots__ = ("v", "capacity", "rev")

        def __init__(self, v, capacity, rev):
            self.v = v            # target node
            self.capacity = capacity
            self.rev = rev        # index of reverse edge

    def __init__(self, n):
        self.n = n
        self.graph = [[] for _ in range(n)]
        self.level = [0] * n
        self.it = [0] * n  # iterators for DFS

    def add_edge(self, u, v, capacity):
        """Add a directed edge u  v with given capacity."""
        forward = self.Edge(v, capacity, len(self.graph[v]))
        backward = self.Edge(u, 0, len(self.graph[u]))
        self.graph[u].append(forward)
        self.graph[v].append(backward)

    def bfs(self, source, sink):
        """Build level graph using BFS."""
        self.level = [-1] * self.n
        q = deque([source])
        self.level[source] = 0

        while q:
            u = q.popleft()
            for edge in self.graph[u]:
                if self.level[edge.v] < 0 and edge.capacity > 0:
                    self.level[edge.v] = self.level[u] + 1
                    q.append(edge.v)

        return self.level[sink] != -1

    def send_flow(self, u, flow, sink):
        """DFS blocking flow."""
        if u == sink:
            return flow

        while self.it[u] < len(self.graph[u]):
            edge = self.graph[u][self.it[u]]

            if edge.capacity > 0 and self.level[edge.v] == self.level[u] + 1:
                pushed = self.send_flow(edge.v, min(flow, edge.capacity), sink)

                if pushed > 0:
                    edge.capacity -= pushed
                    self.graph[edge.v][edge.rev].capacity += pushed
                    return pushed

            self.it[u] += 1

        return 0

    def max_flow(self, source, sink):
        """Compute maximum flow from source to sink."""
        flow = 0
        INF = 10**18

        while self.bfs(source, sink):
            self.it = [0] * self.n
            pushed = self.send_flow(source, INF, sink)
            while pushed:
                flow += pushed
                pushed = self.send_flow(source, INF, sink)

        return flow
````

4. **Recurse and then undo the choice (backtrack)**

### Generic Backtracking

````python
def backtrack(state, choices):
    # 1. If state is a complete solution, process it
    if is_solution(state):
        # process_solution(state)
        return

    # 2. Iterate over all possible choices at this step
    for choice in choices:
        if is_valid(state, choice):
            # 3. Make the choice
            state.append(choice)

            # 4.  Recurse and Undo the choice
            backtrack(state, choices)

            # Undo the choice (backtrack)
            state.pop()
````

### Find all subsets
- Given an array, find all its subsets

````python
# Total results: 2^n subsets.
#Explicit base case: None

# Implicit base case: start == len(nums) (no more choices  recursion stops)
def subsets(nums):
    
    results = []
    solutions = []
    
    def backtrack(index):
        # 1. Check if the current state is a complete solution
        # for subsets, every state is a valid solution
        # skip the checking
        if index == len(nums):
            results.append(solutions[:])
            return 
        # 2. loop through all possible choices
        # 3. Make the choice of skipping and including
        backtrack(index + 1) # skipping
        
        solutions.append(nums[index])

        # 4. recurse and undo
        backtrack(index + 1) # including
        solutions.pop()
            
    backtrack(0)
    return results
````

### Find all permutations
- Find all permutations - order matter

````python
# all permuatations
# Total results: n! permutations.
def permutations(nums):
    results = []

    def backtrack(path):
        # 1. base case
        if len(path) == len(nums):
            results.append(path[:])
            return None
        # 2. loop through all possible choices
        for num in nums:
            if num in path:
                continue
            
            # 3. make the choice
            path.append(num)
            
            # 4. recurse and undo decision
            backtrack(path)
            path.pop()

    backtrack([])

    return results
````

### Queen 8
- placing 8 queens on an 8x8 chessboard so none attack each other (same row, column, or diagonal)

````python
def solve_n_queen(n):
    solutions = []
    board = [-1] * n 
    
    def is_valid(row, col):
        for r in range(row):
            c = board[r] 
            if c == col:  # same column
                return False 
            
            if abs(c - col) == abs(r - row):  # same diagonal
                return False
        return True

    def backtrack(row):
        # 1. check if current state is a complete solution
        if row == n:
            solve = []
            for r in range(n):
                line = ["."] * n 
                line[board[r]] = 'Q'
                solve.append("".join(line))
            solutions.append(solve)
            return None 
        
        # 2. loop through all possible choices
        for col in range(n):
            
            # 3. make a choice if it is valid
            if is_valid(row, col):
                board[row] = col 
                # 4. recurse deeper and undo the choice (backtrack)
                backtrack(row + 1)
                board[row] = -1
        
    backtrack(0)
    return solutions 
    
````

## 18. Greedy
- makes the locally optimal choice at each step with the hope of finding a globally optimal solution. 
- It never reconsiders its past choices, meaning once a decision is made, it's final, making it straightforward to implement but not guaranteed to find the best solution for all problems. 
- This approach is often used in optimization problems like finding the shortest path or coin change problems, provided the problem exhibits the greedy choice property and optimal substructure.

- Structures:
    - Initialization: Start with an empty solution set and a set of candidate items.
        - Selection Rule: While the set of candidate items is not empty and the solution is not complete:
        - Select: Choose the "best" candidate item based on a specific, locally optimal criterion (the greedy choice property).
        - Feasibility Check: Determine if adding this chosen item to the current solution set maintains feasibility (i.e., adheres to all problem constraints).
        - Inclusion: If the item is feasible, add it to the solution set.

    - Update: Remove the chosen item (and potentially other related items) from the candidate set.

    - Termination: Once the loop finishes, the solution set contains the final constructed solution.

- The Greedy Choice Property
    - This property states that a globally optimal solution can be reached by making a locally optimal choice (the greedy choice) at each step. In other words, an optimal solution is guaranteed to contain the greedy choice, so you never have to reconsider this choice later.
- Optimal Substructure
    - This means that an optimal solution to the problem contains within it optimal solutions to subproblems. Once you make the greedy choice, you are left with a smaller subproblem whose optimal solution, when combined with your greedy choice, yields the overall optimal solution.


### When Greedy Algorithms Fail
- It's important to note that if a problem does not possess both the greedy choice property and optimal substructure (like the 0/1 Knapsack Problem or the Traveling Salesperson Problem), a greedy algorithm will only provide an approximate or sub-optimal solution.

### Fractional Knapsack problem
- an optimization problem where you aim to maximize the total value of items placed in a knapsack with a fixed capacity. 
- Unlike the 0/1 knapsack, you can take fractions of items rather than the whole item. 
- The greedy strategy is to pick items in descending order of value-to-weight ratio, taking as much as possible until the knapsack is full. 
- This approach guarantees the maximum total value.

````python
# Each item has a value and a weight
class Item:
    def __init__(self, value, weight):
        self.value = value
        self.weight = weight
        self.ratio = value / weight  # value-to-weight ratio

def fractional_knapsack(items, capacity):
    # Step 1: Sort items by value-to-weight ratio in descending order
    items.sort(key=lambda x: x.ratio, reverse=True)

    solution = []
    total_value = 0
    remaining_capacity = capacity

    for item in items:
        # Step 2: Check feasibility (if item fits entirely)
        if item.weight <= remaining_capacity:
            # Step 3: Make greedy choice
            solution.append((item, 1))  # take whole item
            total_value += item.value
            remaining_capacity -= item.weight
        else:
            # Step 3: Take fraction of the item
            fraction = remaining_capacity / item.weight
            solution.append((item, fraction))
            total_value += item.value * fraction
            remaining_capacity = 0
            break  # knapsack is full

    # Step 4: Return solution
    return total_value, solution
````

### Jump Game
- You are given an integer array nums where each element nums[i] indicates your maximum jump length at that position.

- Return true if you can reach the last index starting from index 0, or false otherwise.

````python
def can_jump(nums):
    # Step 1: Initialize the greedy criterion
    max_reachable = 0
    n = len(nums)

    for i in range(n):
        # Step 2: Check feasibility
        if i > max_reachable:
            return False  # cannot reach this index

        # Step 3: Make greedy choice
        max_reachable = max(max_reachable, i + nums[i])

        # Optional early exit if we can reach the end
        if max_reachable >= n - 1:
            return True

    # Step 4: Return the final decision
    return max_reachable >= n - 1
````

## 19. Dynamic Programming
- DP is primarily a method for solving complex problems by breaking them down into simpler subproblems.

- The core idea of DP relies on two properties: **optimal substructure** (the optimal solution to the overall problem can be constructed from the optimal solutions to its subproblems) and **overlapping subproblems** (the same subproblems are solved repeatedly).

- DP avoids redundant calculations by storing the results of subproblems, a technique known as memoization (top-down DP) or tabulation (bottom-up DP).

- Memoization is a top-down approach where the problem is solved recursively, and a lookup table (e.g., an array or hash map) is used to cache results of subproblems as they are computed.

- Tabulation is a bottom-up approach where you solve the subproblems first (usually starting with the base cases) and store their results in a table, then use these results to compute the solutions to larger subproblems until the main problem is solved.

- Commonly used data structures in DP are arrays or tables (often multi-dimensional) to store the intermediate results and track the optimal solutions for subproblems.

- Classic examples of problems solved efficiently using dynamic programming include the Fibonacci sequence, the Knapsack Problem, the Longest Common Subsequence (LCS), and the minimum path sum in a grid.

### Generic algorithm
- There are usually two methods for solving dynamic programming problems: top-down and bottom-up


````python
# Generic Top-Down (Memoization) Template

def solve_top_down(input_value):
    # Step 1: Initialize the Memoization Table (Cache)
    # The size and type depend on the problem.
    # Use a dictionary or array/list initialized with a 'sentinel' value (like -1 or None)
    # to indicate a result hasn't been computed yet.
    memo = {}

    # Step 2: Define the Recursive Helper Function
    def dp(i):
        # Base Cases (The simplest subproblems)
        if i == 0:
            return 0  # e.g., Fibonacci(0)
        if i == 1:
            return 1  # e.g., Fibonacci(1)

        # Check the Cache (The core of memoization)
        if i in memo:
            return memo[i]

        # Recursive Relation (The transition/recurrence formula)
        # This is where you define the subproblems.
        result = dp(i - 1) + dp(i - 2)  # e.g., Fib(i) = Fib(i-1) + Fib(i-2)

        # Store the result in the cache
        memo[i] = result

        return result

    # Step 3: Call the DP function with the original input
    return dp(input_value)

# Example Call: solve_top_down(6) would return 8 (0, 1, 1, 2, 3, 5, 8)
````

````python
# Generic Bottom-Up (Tabulation) Template

def solve_bottom_up(input_value):
    # Step 1: Initialize the DP Table (usually an array/list)
    # The size is typically (input_value + 1)
    dp_table = [0] * (input_value + 1)

    # Step 2: Initialize the Base Cases
    # Set the values for the simplest subproblems
    if input_value >= 0:
        dp_table[0] = 0  # e.g., Fibonacci(0)
    if input_value >= 1:
        dp_table[1] = 1  # e.g., Fibonacci(1)

    # Step 3: Fill the table iteratively
    # Start from the next step after the base cases
    for i in range(2, input_value + 1):
        # Transition/Recurrence Formula
        # Calculate the current state based on previous, already computed states.
        # e.g., Fib(i) = Fib(i-1) + Fib(i-2)
        dp_table[i] = dp_table[i-1] + dp_table[i-2]

    # Step 4: The final answer is usually the last entry in the table
    return dp_table[input_value]

# Example Call: solve_bottom_up(6) would return 8
````

````python
def fib_tab(n):
    """
    Calculates the n-th Fibonacci number using Tabulation (Bottom-Up DP).
    F(n) = F(n-1) + F(n-2)
    """
    if n <= 1:
        return n

    # 1. Initialize DP Table
    # The table size is (n + 1) to store results from index 0 up to n.
    dp_table = [0] * (n + 1)

    # 2. Define Base Cases in the table
    dp_table[0] = 0
    dp_table[1] = 1

    # 3. Fill the DP Table Iteratively
    # The iterative structure is the core of bottom-up DP.
    for i in range(2, n + 1):
        # State Transition/Recurrence Relation
        # The solution for 'i' depends directly on previously computed solutions.
        dp_table[i] = dp_table[i - 1] + dp_table[i - 2]

    # 4. Return the Final Solution
    return dp_table[n]
````

### Kadanes algorithm
-  a very efficient (linear time, $O(N)$) dynamic programming approach used to find the maximum sum of a contiguous subarray within a one-dimensional array of numbers.
- Kadane's algorithm is essentially a greedy or simplified dynamic programming approach where the state is implicitly defined:
    - DP State: The value of current_max at index $i$ is the maximum sum of a contiguous subarray ending at index i.
    - Recurrence Relation: The transition from state $i-1$ to state $i$ is:$$\text{current\_max}(i) = \max(\text{nums}[i], \text{current\_max}(i-1) + \text{nums}[i])$$
    - This formula captures the core idea: if adding the current number nums[i] to the previous best sum (current_max) results in a smaller value than nums[i] itself, it means the previous subarray was pulling the sum down, and we should start a new subarray at nums[i]. Otherwise, we extend the current subarray.

- By doing this in a single pass (O(N) time complexity), the algorithm efficiently tracks both the current running max and the overall best max.

````python
def kadanes_algorithm(nums):
    """
    Finds the maximum sum of a contiguous subarray using Kadane's algorithm.

    Args:
        nums (list of int): The input array of numbers.

    Returns:
        int: The maximum contiguous subarray sum.
    """
    if not nums:
        # Handle the case of an empty array
        return 0

    # Initialize both variables with the first element of the array
    # This is the starting point for both the overall max and the current subarray max.
    max_so_far = nums[0]
    current_max = nums[0]

    # Iterate through the rest of the array starting from the second element (index 1)
    for i in range(1, len(nums)):
        num = nums[i]

        # 1. Update current_max (The core dynamic programming step)
        # We have two choices for the subarray ending at 'num':
        #   a) Start a new subarray at 'num' (i.e., num itself)
        #   b) Extend the previous subarray (i.e., current_max + num)
        # We pick the one that gives the larger sum.
        current_max = max(num, current_max + num)

        # 2. Update max_so_far
        # Check if the max sum found *ending* at the current position is
        # greater than the overall maximum sum found so far.
        max_so_far = max(max_so_far, current_max)

    return max_so_far
````

````python

# --- Example Usage ---
array1 = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
# The maximum contiguous subarray is [4, -1, 2, 1], with a sum of 6.
print(f"Max sum for Array 1: {kadanes_algorithm(array1)}")  # Output: 6

array2 = [1, 2, 3, -10, 5, 6]
# The maximum contiguous subarray is [5, 6], with a sum of 11.
print(f"Max sum for Array 2: {kadanes_algorithm(array2)}")  # Output: 11
````

- The inner loop runs up to index n - 0 - 1, which is n - 1, covering the entire list.

- Should put a swapped flag to exit early, should the array is already sorted, then bubble sort run O(n)

### Insertion Sort
- The insertion sort algorithm sorts an array by building a sorted sub-array one element at a time. 
- It iterates through the input array, taking each element and inserting it into its correct position within the already sorted portion of the array.
- **Worst case**: O(n^2), **Best case**: O(n^2)
- **DO NOT USE**
- Assume the left part of the array is sorted starting with the first element,
- Then pick the one element at a time from the unsorted part and INSERT that element
- into the correct position in the SORTED left part of the array.
- **Stable Sort**

````python
# in place sorting
def insertion_sort(nums):
    n = len(nums)
    
    # assume the left most array is sorted
    # starting with the second element
    for i in range(1, n):
        key = nums[i]
        j = i
        # going backwards and move it back until it is bigger than the left element
        while j > 0 and key < nums[j - 1]:
            nums[j], nums[j - 1] = nums[j- 1], nums[j]
            
            j -= 1
            
    return nums
````

### Selection Sort
- The BFF of Insertion Sort
- Still part of the trio of useless sort algorithm with Bubble Sort, Insertion Sort and itself.
- The idea is similar to Insertion Sort in that the left part of the array is sorted.
- Now, go through the right side of unsorted array and **SELECT** the minimum
- Swap the minimum with the current index, then continue to **SELECT** each minimum element in order.
- **Worst Case**: O(n^2), **Best Case**: O(n)
- Another DO NOT USE Algorithm.

````python
def selection_sort(nums):
    n = len(nums)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if nums[j] < nums[min_idx]:
                min_idx = j 
        nums[i], nums[min_idx] = nums[min_idx], nums[i]
    return nums
````

### Merge Sort
- an efficient, stable sorting algorithm that follows the divide-and-conquer approach. 
- It works by recursively breaking a list into smaller halves, sorting each half, and then merging the sorted halves back together into a single sorted list. 
- **Use this for SORTING**
- Split the array into 2 halves
- Recurse of each half and sort them
- Merge the sorted chunks together
- **Best, Worst, Average Case**: O(nlogn)



````python
def merge(left, right):
    results = []
    while left and right:
        min_num = left.pop(0) if left[0] < right[0] else right.pop(0)
        results.append(min_num)

    # one of these will be empty, but that's okay
    # extend handles empty arrays just fine
    results.extend(left)
    results.extend(right)
    return results
    
def merge_sort(nums):
    if len(nums) <= 1:
        return nums
    
    mid = len(nums) // 2
    left_half = merge_sort(nums[:mid]) 
    right_half = merge_sort(nums[mid:])
    return merge(left_half, right_half)
    
````

````python
def merge_sort_united(nums):
    n = len(nums)
    if n <= 1:
        return nums
    left_nums = nums[:n//2]
    right_nums = nums[n//2:]

    # recursively divide the nums array
    merge_sort_united(left_nums)
    merge_sort_united(right_nums)

    j = i = k = 0
    # merge the left and right elements together
    while i < len(left_nums) and j < len(right_nums):
        if left_nums[i] < right_nums[j]:
            nums[k] = left_nums[i]
            i += 1
        else:
            nums[k] = right_nums[j]
            j += 1
        k += 1

    # get the remainding elements from either the left or the right
    # while the array is empty, its while loop will not execute
    while i < len(left_nums):
        nums[k] = left_nums[i]
        i += 1
        k += 1

    while j < len(right_nums):
        nums[k] = right_nums[j]
        j += 1
        k += 1

    return nums
````

### Quick Sort
- The quicksort algorithm is a widely used, efficient sorting algorithm that employs a divide-and-conquer strategy. 
- It works by selecting a 'pivot' element from the array and partitioning the other elements into two sub-arrays, according to whether they are less than or greater than the pivot. 
- The sub-arrays are then recursively sorted.
- **USE THIS OR MERGE SORT**
- **Worst Case**: O(n^2) **Best and Average case**: O(nlogn)

````python
def quicksort(arr):
    if len(arr) <= 1:
        return arr 
    else:
        # Choose a pivot (here, the middle element)
        pivot = arr[len(arr) // 2] 
        # Elements smaller than the pivot
        left = [x for x in arr if x < pivot]  
        # Elements equal to the pivot
        middle = [x for x in arr if x == pivot] 
        # Elements greater than the pivot
        right = [x for x in arr if x > pivot]
        # Recursively sort and combine
        return quicksort(left) + middle + quicksort(right)
````

````python
array = [64, 34, 25, 12, 22, 11, 90, 5]
print(quicksort(array))
````

### Quick Select (Not sorting, but related to Quick Sort)
- Efficiently finds the k-th smallest (or largest) element in an unsorted list. 
- It is based on the partitioning idea from Quicksort but avoids sorting the entire list, resulting in an average time complexity of O(n).

````python
def quick_select(nums, k):
    k = len(nums) - k

    def quick_select_recurse(left, right):
        pivot, p = nums[right], left

        for i in range(left, right):
            if nums[i] <= pivot:
                nums[p], nums[i] = nums[i], nums[p]
                p += 1
        nums[p], nums[right] = nums[right], nums[p]

        if p > k:
            return quick_select_recurse(left, p - 1)
        elif p < k:
            return quick_select_recurse(p + 1, right)
        else:
            return nums[p]

    return quick_select_recurse(0, len(nums) - 1)
    
    
````

### Shell Sort
- An in-place comparison sort that is an optimization of Insertion Sort. 
- It works by sorting elements that are far apart and then gradually reducing the gap between compared elements until the gap is 1, at which point it becomes a standard Insertion Sort on an almost sorted array.
- **Worst Case**: O(n^2), **Average Case**: O(n(logn)^2), **Best Case**: O(nlogn)

````python
def shell_sort(nums):
    n = len(nums)
    gap = n // 2
    
    while gap > 0:
        for i in range(gap, n):
            temp = nums[i]
            j = i
            
            while j >= gap and nums[j - gap] > temp:
                nums[j] = nums[j - gap]
                j -= gap
            
            nums[j] = temp
            
        gap //= 2
    return nums
````

### Heap Sort
- The heap sort algorithm sorts an array by first converting it into a max heap, where the largest element is at the root, and then repeatedly extracting the maximum element and placing it at the end of the array. 
- Heapify: The process of converting a binary tree into a heap or restoring the heap property.
- In-place Algorithm: Sorts an array without significant extra memory.
- **Worst, Best, and Average Case**: O(nlogn)

````python
def heapify(nums, n, i):
    """
        helper function for heapify the max-heap
    """
    largest = i
    left = (2 * i) + 1
    right = (2 * i) + 2
    
    if left < n and nums[i] < nums[left]:
        largest = left
    
    if right < n and nums[largest] < nums[right]:
        largest = right 
    
    if largest != i:
        nums[i], nums[largest] = nums[largest], nums[i]
        heapify(nums, n, largest)
        
def heap_sort(nums):
    n = len(nums)
    
    for i in range(n // 2 - 1, - 1, -1):
        heapify(nums, n, i)
    
    for i in range(n - 1, 0, -1):
        nums[i], nums[0] = nums[0], nums[i]
        heapify(nums, i, 0)
        
    return nums
    
````

````python
my_list = [12, 11, 13, 5, 6, 7]
sorted_list = heap_sort(my_list)
print(f"Sorted array: {sorted_list}")
````

### Tim Sort
- combine insertion sort with merge sort
- Timsort uses different sorting algorithms depending on the data's characteristics.
-  It takes advantage of existing order (called "natural runs") in the data to minimize comparisons, allowing it to perform in O(n) time in the best case (when data is already sorted).
- Stable: It preserves the relative order of elements that have equal values.
- Time Complexity: It has an average and worst-case time complexity of O(n log n), which is efficient for large datasets
- It was used as python sorted function before it was replaced with powersort after version 3.11
- **Worst/Average Case**: O(nlogn), **Best Case**: O(n)

````python
import math

MIN_RUN = 32

def calc_min_run(n):
    """Calculates the minimum run length for Timsort."""
    r = 0
    while n >= MIN_RUN:
        r |= n & 1
        n >>= 1
    return n + r


def insertion_sort(arr, left, right):
    """Sorts a small slice of an array using insertion sort."""
    for i in range(left + 1, right + 1):
        j = i
        while j > left and arr[j] < arr[j - 1]:
            arr[j], arr[j - 1] = arr[j - 1], arr[j]
            j -= 1


def merge(arr, l, m, r):
    """Merges two sorted runs in a stable way."""
    # Calculate lengths of the two sub-arrays
    len1, len2 = m - l + 1, r - m
    left_arr, right_arr = [], []

    for i in range(0, len1):
        left_arr.append(arr[l + i])
    for i in range(0, len2):
        right_arr.append(arr[m + 1 + i])

    i, j, k = 0, 0, l

    # Copy data to temp arrays left_arr and right_arr
    while i < len1 and j < len2:
        if left_arr[i] <= right_arr[j]:
            arr[k] = left_arr[i]
            i += 1
        else:
            arr[k] = right_arr[j]
            j += 1
        k += 1

    # Copy remaining elements of left_arr, if any
    while i < len1:
        arr[k] = left_arr[i]
        i += 1
        k += 1

    # Copy remaining elements of right_arr, if any
    while j < len2:
        arr[k] = right_arr[j]
        j += 1
        k += 1


def timsort(arr):
    """Main Timsort algorithm function."""
    n = len(arr)
    min_run = calc_min_run(n)

    # 1. Sort individual runs using insertion sort
    for start in range(0, n, min_run):
        end = min(start + min_run - 1, n - 1)
        insertion_sort(arr, start, end)

    # 2. Merge runs
    size = min_run
    while size < n:
        # Pick starting point of left sub array
        for left in range(0, n, 2 * size):
            # Find ending point of left sub array (mid)
            # Find ending point of right sub array (right)
            mid = min(n - 1, left + size - 1)
            right = min(n - 1, left + 2 * size - 1)

            # Merge sub array arr[left...mid] and arr[mid+1...right]
            if mid < right:
                merge(arr, left, mid, right)

        size = 2 * size

````

### Power Sort
- Powersort refines Timsort's merge policy, the crucial part that decides the order in which sorted subsegments ("runs") are combined. 
- Timsort used a set of heuristics, while Powersort employs a provably near-optimal strategy based on the concept of "powers," which simulates the construction of optimal binary search trees.
- Powersort can be up to 30% faster than Timsort on specific patterns of data where Timsort's original heuristics were suboptimal. 
- For general, randomly ordered data, the speed difference is often tiny, but the change helps avoid "performance blind spots" that Timsort had.


````python
import math

MIN_RUN = 32

def calc_min_run(n):
    """Calculates the minimum run length for Timsort/Powersort."""
    r = 0
    while n >= MIN_RUN:
        r |= n & 1
        n >>= 1
    return n + r

def insertion_sort(arr, left, right):
    """Sorts a small slice of an array using insertion sort."""
    for i in range(left + 1, right + 1):
        j = i
        while j > left and arr[j] < arr[j - 1]:
            arr[j], arr[j - 1] = arr[j - 1], arr[j]
            j -= 1

def merge(arr, l, m, r):
    """Merges two sorted runs in a stable way (same as Timsort merge)."""
    len1, len2 = m - l + 1, r - m
    left_arr, right_arr = arr[l : m + 1], arr[m + 1 : r + 1]
    i, j, k = 0, 0, l

    while i < len1 and j < len2:
        if left_arr[i] <= right_arr[j]:
            arr[k] = left_arr[i]
            i += 1
        else:
            arr[k] = right_arr[j]
            j += 1
        k += 1

    while i < len1:
        arr[k] = left_arr[i]
        i += 1
        k += 1

    while j < len2:
        arr[k] = right_arr[j]
        j += 1
        k += 1

def calculate_node_power(n, b1, e1, b2, e2):
    """
    Calculates a conceptual 'power' based on run midpoints relative to total size 'n'.
    This is a simplification of the complex math used in the C implementation.
    Higher power generally means runs are closer to a virtual balanced tree node.
    """
    n1 = e1 - b1 + 1
    n2 = e2 - b2 + 1
    # Normalized midpoints (a, b)
    a = (b1 + n1 / 2) / n
    b = (b2 + n2 / 2) / n
    
    p = 0
    # Find the highest power of two (depth) where midpoints fall into same segment
    while math.floor(a * (2**p)) == math.floor(b * (2**p)):
        p += 1
    return p

def powersort(arr):
    """Main Powersort algorithm function using a stack for merge management."""
    n = len(arr)
    if n < 2:
        return

    min_run = calc_min_run(n)
    runs = [] # Stack of (start_index, end_index, power) tuples

    # 1. Detect runs and sort small ones
    i = 0
    while i < n:
        start = i
        # This part requires a proper 'FirstRunOf' function to detect natural runs
        # For simplicity in this conceptual code, we create fixed-size 'min_run' chunks
        end = min(start + min_run - 1, n - 1) 
        insertion_sort(arr, start, end)
        
        # In a real powersort, we'd detect the natural run length here
        # and calculate its power relative to its neighbors, then push to stack.

        # The loop below conceptually replaces the stack management loop:
        # The true Powersort stack logic is highly specialized C code, 
        # making a simple Python loop difficult to replicate accurately without
        # significant code complexity beyond the scope of a simple example.
        i = end + 1

    # Because a simplified example cannot fully implement the optimal stack logic
    # without reproducing hundreds of lines of C code, we revert to a bottom-up merge 
    # similar to Timsort's final phase for demonstration.
    # The true Powersort would manage merges optimally on the fly using the NodePower concept.
    
    size = min_run
    while size < n:
        for left in range(0, n, 2 * size):
            mid = min(n - 1, left + size - 1)
            right = min(n - 1, left + 2 * size - 1)
            if mid < right:
                merge(arr, left, mid, right)
        size *= 2


````

### Counting Sort (Trio of Stable sort with Bucket Sort and Radix Sort)
- A non-comparison-based sorting algorithm particularly effective when the range of input values is relatively small compared to the number of elements to be sorted. 
- It works by counting the occurrences of each unique element and using this information to place elements in their correct sorted positions. 
- **Time Complexity**: O(n+k), n = number of elements, k = range of input values from min to max
- **Space Complexity** is also O(n + k) due to the need for both the count array (size k) and the output array (size n).

````python
def counting_sort(arr):
    if not arr:
        return []

    # Find the maximum element in the input array
    max_val = max(arr)

    # Initialize a count array with zeros. Its size will be max_val + 1
    # to accommodate values from 0 to max_val.
    count = [0] * (max_val + 1)

    # Store the frequency of each element in the count array
    for num in arr:
        count[num] += 1

    # Convert the count array into a prefix sum array.
    # Each index now indicates the final position of that element
    # in the sorted output array.
    for i in range(1, len(count)):
        count[i] += count[i - 1]

    # Create an output array of the same size as the input array
    output = [0] * len(arr)

    # Build the output array by iterating through the input array from the end
    # This ensures stability (maintaining the relative order of equal elements)
    for i in range(len(arr) - 1, -1, -1):
        num = arr[i]
        output[count[num] - 1] = num
        count[num] -= 1  # Decrement the count for the current element

    return output
````

### Bucket Sort (Trio of Stable Sort with Coungting Sort and Radix Sort)
- distributing elements into a number of "buckets." Each bucket is then sorted individually, and finally, the sorted elements from all buckets are concatenated to produce the final sorted array. 
- It is particularly efficient when the input data is uniformly distributed over a range. 
- **Best Case**: O(n + k), k = number of buckets
- **Average Case**: O(n)
- **Worst Case**: O(n^2)
- **Space Complexity**: n^2
- Note: A sorting algorithm that uses another sorting, WTF!!!
- **Inner Sorting**: Stable if inner sorting algorithm is stable.
- conditionally stable

````python
def bucket_sort(nums):
    if not nums:
        return []
    
    max_val = max(nums)
    min_val = min(nums)
    
    value_range = max_val - min_val 
    
    if value_range == 0:
        return nums 
    
    nums_bucket = len(nums) // 2 + 1
    
    buckets = [[] for _ in range(nums_bucket)]
    
    for num in nums:
        bucket_index = int((num - min_val) / value_range * (nums_bucket - 1))
        buckets[bucket_index].append(num)
    
    sorted_array = []
    for bucket in buckets:
        bucket.sort()
        sorted_array.extend(bucket)
    
    return sorted_array
    
    
    
````

### Radix Sort - Least Significant Digits (LSD)
- a non-comparison-based sorting algorithm that sorts integers by processing individual digits. 
- It works by repeatedly applying a stable sorting algorithm (like counting sort) to sort the numbers based on each digit, starting from the least significant digit (LSD) and moving to the most significant digit (MSD). 
- **Time Complexity**: O(d * (n + k)), where d is the number of digits, n is the number of elements, and k is the range of values for each digit (10 for decimal numbers). When d and k are constant or small, the complexity is essentially linear, O(n).
- **Space Complexity**: O(n + k), due to the need for temporary output and count arrays in Counting Sort. 


````python
def radix_sort(nums):
    
    def counting_sort(nums, exp):
        n = len(nums) 
        output = [0] * n
        count = [0] * 10
        
        for i in range(n):
            index = nums[i] // exp
            count[index % 10] += 1
        
        for i in range(1, 10):
            count[i] += count[i - 1]
        
        i = n - 1
        while i >= 0:
            index = nums[i] // exp
            output[count[index % 10] - 1] = nums[i]
            count[index % 10] -= 1
            i -= 1
        
        for i in range(n):
            nums[i] = output[i]
        
    
    max_val = max(nums)
    exp = 1
    while max_val // exp > 0:
        counting_sort(nums, exp)
        exp *= 10
    
    return nums
````

### Cube Sort
- A parallel sorting algorithm that utilizes a multi-dimensional array, often conceptually described as a "cube," to sort elements. 
- It's particularly well-suited for parallel processing environments due to its structure.
- **Best Case**: O(n)
- **Average Case**: O(nlogn)
- **Worst Case**: O(nlogn)
- **Space Complexity**: O(n)

````python
def cube_sort(nums, num_participants):
    min_val = min(nums)
    max_val = max(nums)
    if min_val == max_val:
        return sorted(nums) 
    
    range_per_participants = (max_val - min_val) / num_participants
    
    buckets = [[] for _ in range(num_participants)]
    
    for num in nums:
        bucket_index = int((num - min_val) / range_per_participants)
        
        if bucket_index == num_participants:
            bucket_index -= 1
        buckets[bucket_index].append(num)
    
    for i in range(num_participants):
        buckets[i].sort()
        
    sorted_data = []
    for bucket in buckets:
        sorted_data.extend(bucket)
        
    return sorted_data
````

### Modular Arithmetic
- Modular arithmetic is a system of arithmetic for integers, where numbers "wrap around" after reaching a certain value, called the modulus ($n$).  
- It is often called "clock arithmetic" because the most familiar example uses n=12. 
- When we perform addition or multiplication, we only care about the remainder when the result is divided by the modulus. 
- Two integers, $a$ and $b$, are said to be congruent modulo $n$ if they have the same remainder when divided by n, written as $a b mod n.

#### **Properties**
- **Identity** property
    - if **a  b mod m, then  b  a mod m**
- a = b mod m
- Modular arithmetic is compatible with addition, subtraction, and multiplication, meaning you can perform the operation first and then take the remainder, or vice versa.
- If a = b mod m and c = d mod m, then:
    - **Addition:** (a + c) = (b + d) mod m
    - **Subtraction:** (a - c) = (b - d) mod m
    - **Multiplication:** (a * c) = (b * d) mod m
    - **Exponentiation:** a^k = b^k mod m for any non-negative integer k.

- **DIVISION DOES NOT WORK**
#### Division and the Modular Inverse
- Division is the trickiest part. 
- You can only "divide" by multiplying by the modular multiplicative inverse.
- The modular inverse of a modulo m is an integer x such that:
    - a * x = 1 mod m
- The inverse x exists if and only if a and m are relatively prime 
- (i.e., their greatest common divisor is gcd(a, m) = 1).
- This allows you to solve linear congruences like ax = b  mod m. 
- If the inverse a^-1 exists, the solution is: x = a^-1 b mod m.

### Key Theorems (Equations)
- Two famous theorems provide powerful equations for solving complex problems:
    - Fermat's Little Theorem: 
        - If p is a prime number and a is an integer not divisible by p, then:
        - a^(p-1) = 1 mod p

    - Euler's Theorem: 
        - A generalization of Fermat's Little Theorem. If gcd(a, m) = 1, then:
        - a^{\phi(m)} = 1 mod m where \phi(m) is Euler's totient function, which counts the number of positive integers less than m that are relatively prime to m.

### Permutations
- rearrange of elements in a set where changing the order creates a different permutations
- **order matters**: {1, 2, 3} different from {2 ,1, 3}, different from { 3, 2, 1}, etc...
- P(n, r) = (n!) / (n - r)!
- **Real World examples**: number of possible passwords, seating arrangement, ranking/placing 1st, 2nd, 3rd


````python
# LeetCode Permutations
# - Given an integer array nums that may contain duplicates, return all possible subsets (the power set).
# - The solution set must not contain duplicate subsets. Return the solution in any order.
def permutate(nums):
    results = []
    
    def backtrack(current_permutation):
        
        if len(current_permutation) == len(nums):
            results.append(list(current_permutation))
            return None
        
        for num in nums:
            if num not in current_permutation:
                current_permutation.append(num)
                backtrack(current_permutation)
                # need to back track because we are reusing
                # the same array throughout the entire recursion
                # once we add the base case, we need to remove
                # from the list
                current_permutation.pop()
                
    backtrack([])
    return results
````

### Combinations
- a way to select items from a larger set where the **order** of selection doesn't matter, 
- formula is: nCr = n! / r!(n-r)!
- Example: 2 fruits from {Apple, Banana, Cherry}
    - {Apple, Banana}, { Banana, Cherry}, {Apple, Cherry}
    - cannot do {Apple, Banana} and {Banana, Apple}, they are equivalent in combinations

````python
## LeetCode Combiations
# - Given two integers n and k, return all possible combinations of k numbers chosen from the range [1, n].
# - You may return the answer in any order.
def combine(n, k):
    result = []
    current_combination = []
    
    def backtrack(start_num):
        if len(current_combination) == k:
            result.append(list(current_combination))
            return None
        
        if (n - start_num + 1) < (k - len(current_combination)):
            return None
        current_combination.append(start_num)
        backtrack(start_num + 1)
        current_combination.pop()
        
        if start_num + 1 <= n:
            backtrack(start_num + 1)
    
    backtrack(1)
    return result
````

````python
print(combine(4, 2))
print(combine(1, 1))
print(combine(7, 3))
````

### Prime numbers
- a whole number greater than 1 that has only two distinct positive factors: 1 and itself
- **Sieve of Eratosthenes** - a method for finding prime number
    - an ancient, efficient algorithm for finding all prime numbers up to a specified integer (N) by systematically eliminating (sieving out) multiples of each prime, starting with 2, from a list of numbers, leaving only the primes behind.

````python
# generate all prime number up to n
# time complexity is  O(n log(log n))
def sieve_of_erathosthenes(n):
    prime = [True] * (n + 1)
    prime[0] = prime[1] = False 
    
    p = 2
    
    while (p * p  <= n):
        if prime[p] == True:
            for i in range(p *p, n + 1, p):
                prime[i] = False 
        p += 1
        
    prime_numbers = []
    for p in range(2, n + 1):
        if prime[p]:
            prime_numbers.append(p)
    return prime_numbers
````

### Greatest Common Divisor (GCD)
- Euclid Algorithm
-  a simple, effective method for finding the greatest common divisor (GCD) of two numbers. The GCD is the largest number that can divide both numbers without leaving a remainder. 


#### Extended Euclid Algorithm
-  the extended algorithm not only tells you what the greatest common divisor is, but also finds two specific integers (usually called (x) and (y)) that satisfy a special equation called Bzout's identity:
    - (ax+by)=GCD(a,b)

````python
# Time complexity: O(log(min(a,b)))
def gcd_euclidean(a, b):
    while b != 0:
        a, b = b, a % b
    return a
````

````python
print(gcd_euclidean( 48, 60))
````

### Least Common Multiple
- the smallest positive number that is a multiple of two or more given numbers, meaning it's the smallest number perfectly divisible by each of them. 

- a * b = gcd(a, b) * lcm(a, b)
    - gcd(140, 180) * lcm(140, 180) = 140 * 180
    - gcd(140, 180) = 20
    - 20 * lcm(140, 180) = 140 * 180
    - lcm(140, 180) = 140 * 180 / 20 = 1260
    

````python
def lcm(a, b):
    lcm = abs(a * b) // gcd_euclidean(a, b)
    return lcm
````

````python
print(lcm(140, 180))
````

### Gauss formula: (n(n-1)/2)
- The mathematical summation formula for adding up to \(n\) (the sum of the first \(n\) natural numbers) is:

````python
def sum_up_to(n):
    return (n * (n - 1) // 2)
````

````python
print(sum_up_to(3))
print(sum_up_to(13))
print(sum_up_to(23))
````

### Binary Search
- Search through an array-like structure where the elements are sorted

````python
def binary_search(nums, target):
    left = 0
    right = len(nums) - 1
    
    while left <= right:
        mid = left + (right - left) // 2
        
        if nums[mid] == target:
            return target
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
````

````python
# binary search on conditions for
# array of all false follow by all true
# will find the left most true
def binary_search_condition(arr):

    n = len(arr)
    left = 0
    right = n - 1

    while left < right:
        mid = left + ((right - left) // 2)

        if arr[mid]:
            right = mid
        else:
            left = mid + 1

    return left
````

#### pre_order traversal: node -> left -> right

````python
# recursive solution
def pre_order(node):
    if not node:
        return
    print(node)  # process node in some way
    pre_order(node.left)
    pre_order(node.right)
````

````python
# iterative solution
def pre_order(node):
    stack = [node]
    while stack:
        node = stack.pop()
        print(node)  # process node in some way
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
````

#### in_order traversal: left -> node -> right

````python
# recursive solution
def in_order(node):
    if not node:
        return
    in_order(node.left)
    print(node)  # process node in some way
    in_order(node.right)
````

#### post_order traversal: left -> right -> node

````python
# recursive solution
def post_order(node):
    if not node:
        return
    post_order(node.left)
    post_order(node.right)
    print(node)  # process node in some way
````

````python
# iterative solution aka level_order/breadth_first_search
def post_order(node):
    from collections import deque
    queue = deque()
    queue.append(node)
    while queue:
        node = queue.popleft()
        print(node)  # process node in some way
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
````

#### Search in BST

````python
# recusive search for a regular binary tree, children does not
# necessary have any order to it. Just a node with 2 children.
def search_binary_tree(node, target):
    if not node:
        return False  # could return val, true/false or anything else
    if node.val == target:
        return True

    return search_binary_tree(node.left, target) or search_binary_tree(node.right, target)
````

````python

def search_bst(node, target):
    if not node:
        return False
    if node.val == target:
        return True
    elif target < node.val:
        return search_bst(node.left, target)
    else:
        return search_bst(node.right, target)
````

## Union Find 
- Union Find data structure for connecting graphs together

````python
class UnionFind:
    
    def __init__(self, n):
        self.parent = list(range(n))
    
    def find(self, i):
        if self.parent[i] == i:
            return i
        self.parent[i] = self.find(self.parent[i])
        return self.parent[i]
    
    def union(self, i, j):
        root_i = self.find(i)
        root_j = self.find(j)
        
        if root_i != root_j:
            if root_i == 0 or root_j == 0:
                self.parent[root_i] = self.parent[root_j] = 0
            else:
                self.parent[root_i] = root_j
        
````
